 Sequence-to-Sequence Tasks — Using Hugging Face Pipelines for Summarization

STEP 1 — Install and Import Libraries

```python
!pip install transformers torch

from transformers import pipeline
```

Explanation

Pretrained libraries like Hugging Face Transformers are used because:

Training summarization models from scratch needs huge data and time

Pretrained models are already trained on large datasets

 Advantages of Transformer Models

Better context understanding (uses attention mechanism)

Handles long text better than traditional methods

Generates human-like summaries (abstractive)

STEP 2 — Load Summarization Model

```python
summarizer = pipeline("text-generation", model="facebook/bart-large-cnn")
print("Model loaded successfully!")
```

Explanation

Pipeline = ready-made function for NLP tasks

Pretrained model = already trained summarization model

facebook/bart-large-cnn:

Best for news & long text summarization

Produces high-quality abstractive summaries

STEP 3 — Prepare Input Texts

```python
texts = [
"""Artificial Intelligence (AI) is transforming industries by enabling machines
to learn from data and make decisions. AI is used in healthcare, education,
finance, and transportation. However, challenges such as data privacy and
ethical concerns must be addressed.""",

"""The Indian economy has shown steady growth over the past decade.
Government reforms, digitalization, and infrastructure development
have contributed significantly. However, inflation and unemployment
remain key challenges.""",

"""Once upon a time, there was a brave knight who set out on a journey
to rescue a princess. Along the way, he faced many obstacles and learned
valuable lessons about courage and friendship.""",

"""In software engineering, testing is a crucial phase. It ensures that
applications function correctly and meet user requirements. Different
types of testing include unit testing, integration testing, and system testing.""",

"""Hey, did you complete the assignment? Not yet, I was busy yesterday.
We should finish it today before the deadline. Yes, let’s work together and submit it on time."""
]
```

Explanation

Different text types help test model performance

Challenges:

Technical text → complex terms

Conversation → informal language

Stories → narrative flow

STEP 4 — Generate Summaries

```python
for i, text in enumerate(texts):
    summary = summarizer(text, max_length=50, min_length=20, do_sample=False)
    print(f"\nText {i+1} Summary:")
    print(summary[0]['generated_text'])
```

Explanation

Model understands key points and removes unnecessary details

Patterns observed:

Keeps important sentences

Shortens long explanations

Simplifies language

STEP 5 — Control Summary Length

```python
text = texts[0]

short_summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
medium_summary = summarizer(text, max_length=50, min_length=20, do_sample=False)
long_summary = summarizer(text, max_length=80, min_length=30, do_sample=False)

print("Short Summary:\n", short_summary[0]['generated_text'])
print("\nMedium Summary:\n", medium_summary[0]['generated_text'])
print("\nLong Summary:\n", long_summary[0]['generated_text'])
```

Explanation

max_length → maximum words in summary

min_length → minimum words

do_sample → randomness

🔹 Observation

Short → less detail

Medium → balanced

Long → more information

👉 Trade-off:

Short = concise but may lose details

Long = informative but less concise

STEP 6 — Comparison of Text Types (Analysis)

```python
#News texts are summarized very accurately because models are trained on similar data

#Technical texts lose some domain-specific terms

#Narrative texts retain main storyline but lose emotional depth

#Conversational texts are harder to summarize due to informal structure

#Information coverage is highest in structured texts

#Coherence is generally good but may slightly drop in long inputs

#Important details are sometimes skipped in shorter summaries

#Domain-specific words may be simplified or removed

#Model performs best on formal and well-structured text

#Informal and mixed-style text gives less accurate summaries
```
